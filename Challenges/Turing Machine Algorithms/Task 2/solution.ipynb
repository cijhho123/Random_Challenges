## First challenge

Build a TM machine that checks if a given string is a palindrom.

Note that you may use a multi-tape TM, which may greatly facitilate the task.

The alphabet is {a,b}.

See examples of words that are or are-not palindrom under MustPass and MustFail.

```python
turing_machine_name="IsPalindrom"
description="turing machine that cheks if the given string is a palindrom"

def isPalindrom(input_string):
    w=input_string
    wr=w[::-1]
    if w==wr:
        return True
    else:
        return False

input_alphabet = {'a', 'b'}
edge_case_list={"aba","abba"}
controller.add_challenge(turing_machine_name, {'a','b'}, description, isPalindrom, {"aba"})
```

```python
controller.challenges[turing_machine_name].MustPass([("",True), ("bab",True), ("abba",True), ("babbabbabbab",True) ])
controller.challenges[turing_machine_name].MustFail([("aabba",False), ("baabab",False) ])
```

```python
turing_machine_name="IsPalindrom"
description="turing machine that cheks if the given string is a palindrom"

IsPalindrom = TuringMachine(
            states={'q0', 'q_f1', 'q_f2', 'q_acc', 'q_rej', 'q_ret', 'q_eow1', 'q_eow2'},

			input_alphabet= {'a', 'b'} ,

            tape_symbols= {'a', '$', 'b', 'B'},

            transitions={

                ('q0', 'B'): Next('q_acc', 'B', 'R'),
                ('q0', 'a'): Next('q_f1', '$', 'R'),
                ('q0', 'b'): Next('q_f2', '$', 'R'),
                ('q0', '$'): Next('q_acc', '$', 'R'),
                ('q_f1', 'a'): Next('q_f1', 'a', 'R'),
                ('q_f1', 'b'): Next('q_f1', 'b', 'R'),
                ('q_f1', '$'): Next('q_eow1', '$', 'L'),
                ('q_f1', 'B'): Next('q_eow1', 'B', 'L'),
                ('q_f2', 'a'): Next('q_f2', 'a', 'R'),
                ('q_f2', 'b'): Next('q_f2', 'b', 'R'),
                ('q_f2', '$'): Next('q_eow2', '$', 'L'),
                ('q_f2', 'B'): Next('q_eow2', 'B', 'L'),
                ('q_eow1', 'a'): Next('q_ret', '$', 'L'),
                ('q_eow1', 'b'): Next('q_rej', 'B', 'R'),
                ('q_eow1', '$'): Next('q_acc', '$', 'R'),
                ('q_eow2', 'b'): Next('q_ret', '$', 'L'),
                ('q_eow2', 'a'): Next('q_rej', 'B', 'R'),
                ('q_eow2', '$'): Next('q_acc', '$', 'R'),
                ('q_ret', 'a'): Next('q_ret', 'a', 'L'),
                ('q_ret', 'b'): Next('q_ret', 'b', 'L'),
                ('q_ret', '$'): Next('q0', '$', 'R'),
            },
            initial_state='q0',
            accept_states={'q_acc'},
            reject_states={'q_rej'}
        )

controller.update_turing_machine("IsPalindrom", IsPalindrom)


input_alphabet = {'a', 'b'}
edge_case_list={"aba","abba"}
```

## Second challenge

Build a TM that receives a unary string such as "11111111", and returns a unary string of length half the length of the given string.

Note that here you need to implement a TM that conputes a function from strings to strings (rather than defines a language).

Use a multi-tape TM. Recall that the input is on the first tape (indexed 0) and the output is on the last tape.

The alphabet is {1}.

See examples of words valid input-output pairs under MustPass.



```python
turing_machine_name="DivideByTwo"
description="turing machine that gets a unary string and returns a unary string of half the length of the input"

def divideByTwo(unary_string):
    n=len(unary_string)
    m=n//2
    res = '1'*m
    return res
input_alphabet = {'1'}
edge_case_list={"111", "1111", ""}

DivideByTwo = MultiTapeTuringMachine(
    states={'q1','q2', 'q_acc'},
    input_alphabet={'1'},
    tape_alphabet={'1', 'B',''},
    transition_function= {

    ('q1', '1','B'): MultiNext('q2', '1', '', 'R', 'S'),
    ('q1', 'B','B'): MultiNext('q_acc', '', '', 'S', 'S'),


    ('q2', '1',''): MultiNext('q1', '1', '1', 'R', 'R'),
    ('q2', 'B',''): MultiNext('q_acc', '', '', 'S', 'S'),
},
    start_state='q1',
    accept_state={'q_acc'},
    reject_state={},
    num_tapes=2
)

controller.update_turing_machine('DivideByTwo', DivideByTwo, "tape")

inputs = ['', '']
controller.run_turing_machine("DivideByTwo", inputs)
```

```python
controller.add_challenge(turing_machine_name, {'1'}, description, divideByTwo, {"1111", "11"})
controller.challenges[turing_machine_name].MustPass([("",""), ("1",""), ("11","1"), ("111","1"), ("1111","11"),  ("11111","11"), ("11111","11")])
```

## Third challenge
Build a TM that receives a unary string such as "11111111", and returns a unary string of length equal to the log on base 2 of the length of the given string.

Note that here you need to implement a TM that conputes a function from strings to strings (rather than defines a language).

Use a multi-tape TM. Recall that the input is on the first tape (indexed 0) and the output is on the last tape.

The alphabet is {1}.

See examples of words valid input-output pairs under MustPass.

```python
import math
turing_machine_name="LogBaseTwo"
description="turing machine that gets a unary string and returns a unary string of length log 2 of the length of the input string"

def logBaseTwo(unary_string):
    n=len(unary_string)
    if n==0:
        return ""
    m=math.log2(n)
    res = '1'*int(m)
    return res
input_alphabet = {'1'}
edge_case_list={"111", "1111", ""}

logBaseTwoTM = MultiTapeTuringMachine(
    states={'q_0', 'q2+', 'q_check', 'q#', 'q_ret', 'q_end', 'q_cls', 'q_acc', 'q_roll'},
    input_alphabet={'1'},
    tape_alphabet={'1', 'B', '', '#', "!", "*"},
    transition_function= {

    ('q2+', 'B',''): MultiNext('q_acc', '#', '', 'L', 'R'),
    ('q_check', '#',''): MultiNext('q#', '*', '', 'R', 'S'),
    ('q_check', '#','B'): MultiNext('q#', '*', '', 'R', 'S'),
    ('q_check', '!',''): MultiNext('q_end', '!', '1', 'R', 'R'),
    ('q_0', '1','B'): MultiNext('q2+', '#', '', 'R', 'S'),
    ('q_0', 'B','B'): MultiNext('q_acc', '', '', 'R', 'S'),
    ('q2+', '1',''): MultiNext('q_check', '#', '1', 'L', 'R'),
    ('q#', '#',''): MultiNext('q#', '#', '', 'R', 'S'),
    ('q#', '!',''): MultiNext('q#', '!', '', 'R', 'S'),
    ('q#', '1',''): MultiNext('q_ret', '!', '', 'L', 'S'),
    ('q_roll', 'B',''): MultiNext('q_acc', '#', '', 'L', 'R'),
    ('q_roll', '#',''): MultiNext('q#', '*', '', 'R', 'S'),
    ('q_roll', '1','B'): MultiNext('q#', '*', '', 'R', 'S'),
    ('q#', 'B',''): MultiNext('q_acc', '!', '', 'L', 'S'),
    ('q_ret', '!',''): MultiNext('q_ret', '!', '', 'L', 'S'),
    ('q_ret', '#',''): MultiNext('q_ret', '#', '', 'L', 'S'),
    ('q_ret', '*',''): MultiNext('q_check', '*', '', 'R', 'S'),
    ('q_end', '!','B'): MultiNext('q_end', '!', '', 'R', 'S'),
    ('q_end', '1','B'): MultiNext('q_cls', '1', '', 'L', 'S'),
    ('q_end', 'B','B'): MultiNext('q_acc', 'B', '', 'R', 'S'),
    ('q_end', '!',''): MultiNext('q_end', '!', '', 'R', 'S'),
    ('q_end', '1',''): MultiNext('q_cls', '1', '', 'L', 'S'),
    ('q_end', 'B',''): MultiNext('q_acc', 'B', '', 'R', 'S'),
    ('q_cls', '!',''): MultiNext('q_cls', '#', '', 'L', 'S'),
    ('q_cls', 'B',''): MultiNext('q_check', 'B', '', 'R', 'S'),
    ('q_cls', '#',''): MultiNext('q_check', '#', '', 'S', 'S'),
    ('q_cls', '*',''): MultiNext('q_cls', '#', '', 'L', 'S'),

},
    start_state='q_0',
    accept_state={'q_acc'},
    reject_state={},
    num_tapes=2
)

controller.add_challenge(turing_machine_name, {'1'}, description, logBaseTwo, {"111"})
controller.update_turing_machine("LogBaseTwo", logBaseTwoTM, "tape")


inputs = ['', '']
controller.run_turing_machine("LogBaseTwo", inputs)

```
